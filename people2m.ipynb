```python
import numpy as np
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq #ใช้pythonใช้pyarrow
import pyarrow.dataset as ds
```

```python
table = pa.Table.from_pandas(dfp)
print(table)
print("Saving to parquet files...")

# Drop result and table columns
table = table.drop(["User Id", "Email","Phone"])
print(table)
```

```python
table.to_pandas()
```

ให้ตรวจสอบตำแหน่งงานที่มีในรายการว่ามีทั้งหมดกี่ตำแหน่งงาน แต่ละตำแหน่งงานมีจำนวนกี่คน

```python
import pyarrow.compute as pc
grouped_table = table.group_by("Job Title").aggregate(
  [("Index",
  "count",  #count #max #min # sum

  pc.CountOptions(mode="only_valid"))]
)

grouped_table.to_pandas()
```

กรองข้อมูลโดยแสดงเฉพาะข้อมูลของคนที่เป็นเพศชายและมีตำแหน่งเป็น "Writer" เท่านั้น

```python
expr = pc.field("Job Title") == "Writer"
table.filter(expr).to_pandas()
```

```python
expr2 = pc.field("Sex") == "Male"
```

```python
table.filter(expr2).to_pandas()
```

```python
filtered_df = table.sort_by([("Sex", "ascending")]).filter(expr2| expr).to_pandas()
```

```python
filtered_df = table.sort_by([("Job Title", "ascending")]).filter((expr &expr2) ).to_pandas()
```
